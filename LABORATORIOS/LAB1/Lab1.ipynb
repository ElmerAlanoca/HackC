```python
#3. Imprimir el valor del píxel con coordenadas x= 234 y y= 55.
pixel = image[234,55]
print(pixel)
```

```python
#4. Encontrar los píxeles más oscuros y más claros de la imagen.
size = image.shape
min = np.amin(image)
ubic_min = np.where(image == np.amin(image))
print(min)
print(ubic_min)
```

```python
prueba = np.full(size,255)
prueba[ubic_min[0],ubic_min[1]] = min

plt.imshow(prueba)
```

```python
max = np.amax(image)
ubic_max = np.where(image == np.amax(image))
print(max)
print(ubic_max)
```

```python
prueba = np.full(size,0)
prueba[ubic_max[0],ubic_max[1]] = max

plt.imshow(prueba)
```

```python
    #prueba numero de repeticiones
size = (8,8)
img = np.random.rand(8,8)*255
img = img.astype(int)
print(img)
plt.imshow(img)
```

```python
size = img.shape
print(size)
```

```python
rep0 = np.zeros(255)
for i in range(size[0]-1):
  rep = np.bincount(img[i])
  rep1 = np.zeros(255-rep.size)
  rep = np.concatenate((rep, rep1), axis=0)
  rep0 = rep0 + rep
print(rep0)
```

```python
maximo = np.amax(rep0)
print(maximo)

for i in range(len(rep0)):
    if rep0[i] == maximo:
      print(i, end=' ')
```

```python
#4. Dada una imagen en escala de grises, encuentre el valor de píxel que más se repite.
uploaded = files.upload()
img = cv2.imread('Starman2.jpg') 
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)

plt.imshow(img)
```

```python
size = img.shape
print(size)
```

```python
rep0 = np.zeros(255+1)
for i in range(size[0]-1):
  rep = np.bincount(img[i])
  rep1 = np.zeros(255+1-rep.size)
  rep = np.concatenate((rep, rep1), axis=0)
  rep0 = rep0 + rep
print(rep0)
```

```python
maximo = np.amax(rep0)
print(maximo) #las veces que se repite

for i in range(len(rep0)):  #los pixeles que se repiten
    if rep0[i] == maximo:
      print(i, end=' ')
```
